```python
import numpy as np
import matplotlib.pyplot as plt

xs = [0, 1, 2, 2]
ys = [0, 0, 1, 2]

x = np.array(xs)
y = np.array(ys)

a1, a0 = np.polyfit(x, y, 1)

x_linea = np.linspace(-0.5, 2.5, 100)
y_linea = a1 * x_linea + a0

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='red', s=100, label='Datos: (0,0), (1,0), (2,1), (2,2)', zorder=5)
plt.plot(x_linea, y_linea, color='blue', linewidth=2, 
         label=f'Recta de Ajuste: y = {a1:.2f}x {"+" if a0 > 0 else "-"} {abs(a0):.2f}')

plt.title('Ajuste de Mínimos Cuadrados', fontsize=14, fontweight='bold')
plt.xlabel('Variable Independiente (X)', fontsize=12)
plt.ylabel('Variable Dependiente (Y)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(np.arange(-1, 4, 1))
plt.yticks(np.arange(-1, 4, 1))
plt.legend(fontsize=11)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

xs = [3.38, 0.35, 2.07, -0.45, -0.55, -1.06, -2.77, 3.08, -3.98, -5.0, -3.18, -1.96, 2.37, -1.66, 4.09]
ys = [-0.04, 1.28, 0.65, 2.18, 1.70, 1.96, 3.36, 0.18, 3.35, 4.16, 2.95, 2.34, 0.38, 1.75, -1.06]

x = np.array(xs)
y = np.array(ys)

n = len(x)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_x2 = np.sum(x**2)
sum_xy = np.sum(x * y)

print(f"Eq 1 (db): {n}*b + ({sum_x})*m = {sum_y}")
print(f"Eq 2 (dm): ({sum_x})*b + ({sum_x2})*m = {sum_xy}")
print("-" * 55)

m, b = np.polyfit(x, y, 1)

print(f"b: {b}")
print(f"m: {m}")
print(f"y = {m}x + {b}")

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='red', s=60)

x_linea = np.linspace(min(x) - 1, max(x) + 1, 100)
y_linea = m * x_linea + b

plt.plot(x_linea, y_linea, color='blue', linewidth=2.5)

plt.title('Método de Mínimos Cuadrados con una funcion lineal', fontsize=14, fontweight='bold')
plt.xlabel('Variable X', fontsize=12)
plt.ylabel('Variable Y', fontsize=12)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

xs = [3.38, 0.35, 2.07, -0.45, -0.55, -1.06, -2.77, 3.08, -3.98, -5.0, -3.18, -1.96, 2.37, -1.66, 4.09]
ys = [-0.04, 1.28, 0.65, 2.18, 1.70, 1.96, 3.36, 0.18, 3.35, 4.16, 2.95, 2.34, 0.38, 1.75, -1.06]

x = np.array(xs)
y = np.array(ys)

n = len(x)
sum_x = np.sum(x)
sum_x2 = np.sum(x**2)
sum_x3 = np.sum(x**3)
sum_x4 = np.sum(x**4)

sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2y = np.sum((x**2) * y)

print(f"Eq 1: {n}*a0 + ({sum_x})*a1 + ({sum_x2})*a2 = {sum_y}")
print(f"Eq 2: ({sum_x})*a0 + ({sum_x2})*a1 + ({sum_x3})*a2 = {sum_xy}")
print(f"Eq 3: ({sum_x2})*a0 + ({sum_x3})*a1 + ({sum_x4})*a2 = {sum_x2y}")

a2, a1, a0 = np.polyfit(x, y, 2)

print(f"a0: {a0}")
print(f"a1: {a1}")
print(f"a2 : {a2}")
print(f"Ecuación: y = {a2}x² + {a1}x + {a0}")

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='red', s=60)

x_linea = np.linspace(min(x) - 1, max(x) + 1, 100)
y_linea = a2 * x_linea**2 + a1 * x_linea + a0

plt.plot(x_linea, y_linea, color='blue', linewidth=2.5)

plt.title('Regresión Cuadrática - Método de Mínimos Cuadrados', fontsize=14, fontweight='bold')
plt.xlabel('Variable X', fontsize=12)
plt.ylabel('Variable Y', fontsize=12)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

xs = [3.38, 0.35, 2.07, -0.45, -0.56, -1.06, -2.78, 3.08, -3.99, -5.00, -3.18, -1.97, 2.37, -1.67, 4.09, -4.60, 2.68, 2.78, -3.79, -1.87]
ys = [15.65, -11.20, -4.00, 18.03, 7.94, 15.32, -4.40, 1.39, -11.92, -90.24, 6.92, 28.35, -2.41, -5.47, 35.91, -55.53, 0.77, 4.79, -27.05, 2.85]

x = np.array(xs)
y = np.array(ys)

a3, a2, a1, a0 = np.polyfit(x, y, 3)


y_pred = a3 * x**3 + a2 * x**2 + a1 * x + a0

errores_abs = np.abs(y - y_pred)

error_promedio = np.mean(errores_abs)
error_std = np.std(errores_abs)

idx_max = np.argmax(errores_abs)
idx_min = np.argmin(errores_abs)

x_max, y_max, err_max = x[idx_max], y[idx_max], errores_abs[idx_max]
x_min, y_min, err_min = x[idx_min], y[idx_min], errores_abs[idx_min]

print("--- ANÁLISIS DE ERROR (Interpolación Cúbica) ---")
print(f"Error Absoluto Promedio: {error_promedio:.2f}")
print(f"Desviación Estándar del Error: {error_std:.2f}")
print("-" * 50)
print(f"Punto con Error MÁXIMO: ({x_max}, {y_max}) -> Error de {err_max:.2f}")
print(f"Punto con Error MÍNIMO: ({x_min}, {y_min}) -> Error de {err_min:.2f}")

plt.figure(figsize=(12, 7))

plt.scatter(x, y, color='red', s=50, label='Datos reales', zorder=5)

x_linea = np.linspace(min(x) - 0.5, max(x) + 0.5, 100)
y_linea = a3 * x_linea**3 + a2 * x_linea**2 + a1 * x_linea + a0
plt.plot(x_linea, y_linea, color='blue', linewidth=2, label='Ajuste Cúbico')

plt.scatter(x_max, y_max, color='black', s=150, marker='s', zorder=10, 
            label=f'Error Máx ({err_max:.2f})')
plt.plot([x_max, x_max], [y_max, y_pred[idx_max]], color='black', linestyle=':', linewidth=2)

plt.scatter(x_min, y_min, color='gold', edgecolor='black', s=300, marker='*', zorder=10, 
            label=f'Error Mín ({err_min:.2f})')
plt.plot([x_min, x_min], [y_min, y_pred[idx_min]], color='gold', linestyle=':', linewidth=2)

plt.title('Regresión Cúbica y Análisis de Errores', fontsize=14, fontweight='bold')
plt.xlabel('Variable X', fontsize=12)
plt.ylabel('Variable Y', fontsize=12)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=11)

plt.show()
```
